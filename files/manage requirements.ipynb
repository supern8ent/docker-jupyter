```python
def make_diff(a, b):
    al = [x.strip() for x in a.split('\n') if x.strip()]
    bl = [x.strip() for x in b.split('\n') if x.strip()]
    return '\n'.join([x for x in bl if x not in al])
```

```python
layers_requirements = list()
previous = ''
for name, packages, freeze in layers_freeze:
    this_libs = make_diff(previous, freeze)
    layers_requirements.append((name, packages, this_libs))
    previous = previous + '\n' + this_libs
```
